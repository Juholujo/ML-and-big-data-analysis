```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Прочитаем данные
df = pd.read_csv('Hotel Reservations.csv')

# Укажем интересующие нас категориальные признаки,
# по которым хотим смотреть распределение таргета
features = [
    'type_of_meal_plan',
    'required_car_parking_space',
    'room_type_reserved',
    'market_segment_type',
    'repeated_guest',
    'no_of_special_requests'
]

# Настроим размер сетки: 2 строки x 3 столбца = 6 графиков
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))

for idx, col in enumerate(features):
    # Строим countplot: по оси X — признак col,
    # а цветовая группировка (hue) — booking_status
    sns.countplot(
        data=df,
        x=col,
        hue='booking_status',
        ax=axes[idx // 3, idx % 3]
    )

    # Дадим заголовок
    axes[idx // 3, idx % 3].set_title(f'Распределение статуса по признаку "{col}"')

# Для красоты поджимаем вёрстку
plt.tight_layout()
plt.show()
```
